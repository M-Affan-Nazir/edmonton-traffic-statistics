# Heat Map for Correlation:

```python
corr_collision = annual_collision_report_raw.corr()
plt.figure(figsize=(15, 15))
plt.imshow(corr_collision, interpolation='nearest')
plt.xticks(range(len(corr_collision.columns)), corr_collision.columns, rotation=90)
plt.yticks(range(len(corr_collision.columns)), corr_collision.columns)
plt.colorbar()
plt.title('Correlation Heatmap - Collision')
plt.tight_layout()
plt.show()
```

```python
corr_traffic_volume = average_annual_weekday_traffic_volume_raw.select_dtypes(include=[np.number]).corr()
plt.figure(figsize=(10, 8))
plt.imshow(corr_traffic_volume, interpolation='nearest')
plt.xticks(range(len(corr_traffic_volume.columns)), corr_traffic_volume.columns, rotation=90)
plt.yticks(range(len(corr_traffic_volume.columns)), corr_traffic_volume.columns)
plt.colorbar()
plt.title('Correlation Heatmap - Traffic Volume')
plt.tight_layout()
plt.show()
```
